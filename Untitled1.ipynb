```python
import pandas as pd

import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
# метод pandas get_dummies
one_hot_data = pd.get_dummies(data['whoAmI'])
one_hot_data.head()
```

```python
one_hot_data = pd.DataFrame({'human': data['whoAmI'] == 'human', 'robot': data['whoAmI'] == 'robot'}).astype(int)

one_hot_data.head()
```
