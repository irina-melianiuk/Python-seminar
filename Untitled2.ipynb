```python
import pandas as pd

import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
# Для кодировки категориальных данных можно использовать метод pandas get_dummies
pd.get_dummies(data['whoAmI']).head()
```
